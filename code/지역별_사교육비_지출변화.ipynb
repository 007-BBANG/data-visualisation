# 지역별 사교육비 지출 변화
시도별 1인당 월평균 사교육비

출처: 괜찮은 뉴스 - 사교육비의 모든 것 #20. 시도별 학교급별 (https://www.nextplay.kr/news/articleView.html?idxno=7257)

```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

data = pd.read_csv('/content/drive/MyDrive/2025 공공데이터 활용대회 데이터/지역별 학생 1인당 사교육비(고등학생).csv')
data
```

```python
data.drop('전체', axis=1, inplace=True)
data = data.set_index('연도')
data
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# 선 그래프 그리기
plt.plot(data.index, data["서울"], marker="o", label="서울")
plt.plot(data.index, data["부산"], marker="o", label="부산")
plt.plot(data.index, data["대구"], marker="o", label="대구")
plt.plot(data.index, data["인천"], marker="o", label="인천")
plt.plot(data.index, data["광주"], marker="o", label="광주")
plt.plot(data.index, data["대전"], marker="o", label="대전")
plt.plot(data.index, data["울산"], marker="o", label="울산")
plt.plot(data.index, data["세종"], marker="o", label="세종")

# 그래프 제목 및 축 레이블
plt.title("지역별 1인당 월평균 사교육비(고등학교)", fontsize=14)
plt.xticks(data.index, rotation=45)
plt.xlabel("연도", fontsize=12)
plt.ylabel("사교육비 (만원)", fontsize=12)
plt.legend()
plt.grid(True)

# 그래프 출력
plt.show()
```
