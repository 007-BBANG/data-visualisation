```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/한국언론진흥재단_뉴스빅데이터_연도 및 과목별 언론의 수능난이도 평가_20221231.csv', encoding='utf-8')
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
!pip install koreanize-matplotlib
import koreanize_matplotlib

pivot_df = df.pivot(index="연도", columns="과목", values="언론의 추정 난이도 평균")

# 막대 너비 설정
bar_width = 0.25
x_indexes = np.arange(len(pivot_df.index))

# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.bar(x_indexes - bar_width, pivot_df["국어"], width=bar_width, color='red', label="국어")
plt.bar(x_indexes, pivot_df["수학"], width=bar_width, color='blue', label="수학")
plt.bar(x_indexes + bar_width, pivot_df["영어"], width=bar_width, color='green', label="영어")

# X축 설정
plt.xticks(ticks=x_indexes, labels=pivot_df.index)

# 그래프 제목 및 레이블
plt.xlabel("연도")
plt.ylabel("언론의 추정 난이도 평균")
plt.title("연도별 수능 난이도 (과목별)")
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)  # 기준선 추가
plt.legend()
plt.grid(axis='y', linestyle="--", alpha=0.7)

# 그래프 보여주기
plt.show()
```
