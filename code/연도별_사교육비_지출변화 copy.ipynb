# 연도별 사교육비 지출 변화
교육단계별 학생 1인당 월평균 사교육비

※ 가구소득별 데이터 제외하고 사용

출처: 지표누리 국가발전지표 - 학생 1인당 사교육비
 (https://www.index.go.kr/unify/idx-info.do?idxCd=4244)

```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

data = pd.read_csv('/content/drive/MyDrive/2025 공공데이터 활용대회 데이터/연도별 학생 1인당 사교육비(교육 단계별).csv')
data
```

```python
# 인덱스 열 변경
data = data.set_index('연도')
data
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')

# 그래프 크기 설정
plt.figure(figsize=(10, 6))

# 선 그래프 그리기
plt.plot(data.index, data["전체"], marker="o", label="전체")
plt.plot(data.index, data["초등학교"], marker="s", label="초등학교")
plt.plot(data.index, data["중학교"], marker="^", label="중학교")
plt.plot(data.index, data["고등학교"], marker="*", label="고등학교")

# 그래프 제목 및 축 레이블
plt.title("연도별 교육 단계별 1인당 월평균 사교육비", fontsize=14)
plt.xticks(data.index, rotation=45)
plt.xlabel("연도", fontsize=12)
plt.ylabel("사교육비 (만원)", fontsize=12)
plt.legend()
plt.grid(True)

# 그래프 출력
plt.show()
```
