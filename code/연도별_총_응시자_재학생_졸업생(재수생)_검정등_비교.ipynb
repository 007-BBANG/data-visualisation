```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/한국교육과정평가원_대학수학능력시험 연도별 접수현황_20241231.csv', encoding='cp949')
```

```python

columns_to_drop = ['기간(1)', '기간(2)', '남', '여', '인문', '자연', '예체능', '3개 영역 이하 응시수수료', '4개 영역 응시수수료', '5개 영역 응시수수료', '6개 영역 응시수수료']

df = df.drop(columns=columns_to_drop, errors='ignore')

df
```

```python
df.isnull().sum()
```

```python
import numpy as np

Q1 = df["총 계"].quantile(0.25)  # 1사분위수
Q3 = df["총 계"].quantile(0.75)  # 3사분위수
IQR = Q3 - Q1

# 정상 범위: [Q1 - 1.5*IQR, Q3 + 1.5*IQR]
df = df[~((df["총 계"] < (Q1 - 1.5 * IQR)) | (df["총 계"] > (Q3 + 1.5 * IQR)))]
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
!pip install koreanize-matplotlib
import koreanize_matplotlib

import seaborn as sns

df["재수생 비율"] = (df["졸업생"] / df["총 계"]) * 100

plt.figure(figsize=(12, 6))
sns.lineplot(x=df["학년도"], y=df["재수생 비율"], marker="o", linewidth=2, color="b")

plt.xlabel("학년도")
plt.ylabel("재수생 비율 (%)")
plt.title("연도별 재수생 비율 변화")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 필요한 컬럼만 선택
df = df[["학년도", "총 계", "재학생", "졸업생", "검정등"]]

# 데이터 타입 변환 (혹시 숫자가 문자열이면 변환)
df = df.astype({"총 계": int, "재학생": int, "졸업생": int, "검정등": int})

# 그래프 크기 설정
plt.figure(figsize=(14, 6))

# x축 라벨 (학년도)
x = np.arange(len(df["학년도"]))

# 막대 너비 설정
bar_width = 0.2

# 각 그룹의 막대 그래프 그리기
plt.bar(x - bar_width * 1.5, df["총 계"], width=bar_width, label="총 계", color="blue")
plt.bar(x - bar_width * 0.5, df["재학생"], width=bar_width, label="재학생", color="green")
plt.bar(x + bar_width * 0.5, df["졸업생"], width=bar_width, label="졸업생(재수생)", color="red")
plt.bar(x + bar_width * 1.5, df["검정등"], width=bar_width, label="검정등", color="purple")

# x축 학년도 설정
plt.xticks(x, df["학년도"], rotation=45)

# 제목 및 축 라벨
plt.xlabel("학년도")
plt.ylabel("응시자 수")
plt.title("연도별 총 응시자, 재학생, 졸업생(재수생), 검정등 비교")

# 범례 추가
plt.legend()

# 격자 추가
plt.grid(axis="y", linestyle="--", alpha=0.7)

# 그래프 출력
plt.show()
```
